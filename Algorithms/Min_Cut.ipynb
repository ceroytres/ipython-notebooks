# Karger's algorithm for minimum cuts

```python
import random

class undirected_Graph(object):
    
    def __init__(self,graph_dict={}):
        
        self._graph_dict=graph_dict
    
    def add_node(self,node):
        if node not in self._graph_dict:
            self._graph_dict[node]=[]
    
    def add_edge(self,edge):
        node1=edge[0]
        node2=edge[1]
        
        self._graph_dict[node1].append(node2)
        self._graph_dict[node2].append(node1)
        
    def remove_edge(self,edge):
        node1=edge[0]
        node2=edge[1]
        
        self._graph_dict[node1].remove(node2)
        self._graph_dict[node2].remove(node1)
    
    def get_edges(self):
        edges=[]
        for n in self._graph_dict.keys():
            for v in self._graph_dict[n]:
                edges.append((n,v))
                
            if len(self._graph_dict[n])==0:
                edges.append((n,[]))
        return edges

    def get_nodes(self):
        return self._graph_dict.keys()
    
    def build_supernode(self,u,v):
        temp=(u,v)
        node= tuple(j for i in temp for j in (i if isinstance(i, tuple) else (i,)))
        
        #print "u:", (u)
        #print "v: ", v
        #print "(u,v)", node
        u_nbr=self._graph_dict[u]
        
        while v in u_nbr: 
            u_nbr.remove(v)
        
        v_nbr=self._graph_dict[v]
        while u in v_nbr:
            v_nbr.remove(u)
            
        del self._graph_dict[u]
        del self._graph_dict[v]
        
        for n in u_nbr:
            self._graph_dict[n]=map(lambda x: node if x==u else x ,self._graph_dict[n])
        
        for n in v_nbr:
            self._graph_dict[n]=map(lambda x: node if x==v else x ,self._graph_dict[n])
        
        self._graph_dict[node]=v_nbr+u_nbr
        
def parse_paj(filepath):
    G=undirected_Graph()
    state=True
    with open(filepath,'r') as f:
        for line in f:

            if "*Edges" in line:
                state=False
                continue
            elif "*Vertices" in line:
                continue
                
            if state:    
                node=int(line.split(' ')[0])
                G.add_node(node)
            else:
                node1=int(line.split(' ')[0])
                node2=int(line.split(' ')[1])
                G.add_edge([node1,node2])
                
    return G

 


def rand_contraction(G):
    num_V=len(G.get_nodes())
    while num_V>2:
        edges=G.get_edges()
        num_e=len(edges)
        
        idx=random.randrange(num_e)
        u,v=edges[idx]
        G.build_supernode(u,v)
        num_V=len(G.get_nodes())
        
    return G,(len(G.get_edges())/2)
    
    
    
        
#G=parse_paj("./karate.paj") 

G=undirected_Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_edge([1,2])
G.add_edge([2,3])
G.add_edge([3,4])
G.add_edge([1,4])
G.add_edge([2,4])



P,cut_size=rand_contraction(G)

for e in P.get_edges():
    print "edge: ", e 
print "nodes ", P.get_nodes()
print "Cut Size %d" %(cut_size)

```
